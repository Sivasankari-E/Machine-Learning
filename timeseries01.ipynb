```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
samples = np.random.normal(loc=0, scale=1, size=1000)
samples.size
```

```python
plt.figure(figsize=(10, 6))
```

```python
plt.plot(samples, label='Standard Normal Samples',)
plt.title('Time Series of Standard Normal Samples')
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
plt.hist(samples, label='Standard Normal Samples')
plt.title('Time Series of Standard Normal Samples')
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```
